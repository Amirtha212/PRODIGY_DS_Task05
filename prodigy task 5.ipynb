```python
plt.figure(figsize=(9, 5))
plt.scatter(df['Longitude'], df['Latitude'], alpha=0.5, color='red')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Accident Hotspots')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(9, 5))
sns.countplot(x='Road_Type', data=df)
plt.title('Distribution of Accidents by Road Conditions')
plt.xlabel('Road_Type')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(9, 5))
sns.countplot(x='Weather_Conditions', data=df)
plt.title('Distribution of Accidents by Weather Conditions')
plt.xlabel('Weather Conditions')
plt.ylabel('Number of vehicles')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(9, 5))
sns.histplot(df['Time'], bins=24, kde=True)
plt.title('Distribution of Accidents by Time of Day')
plt.xlabel('Time')
plt.ylabel('Number of vehicles')
plt.xticks(rotation=45)
plt.show()
```
